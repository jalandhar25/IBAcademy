## Sub-matrix Sum Queries:
Given a matrix of integers A of size N x M and multiple queries Q, for each query find and return the submatrix sum.

Inputs to queries are top left (b, c) and bottom right (d, e) indexes of submatrix whose sum is to find out.

Note: Rows are numbered from top to bottom and columns are numbered from left to right. Sum may be large so return the answer mod 10^9 + 7. 

### Input Format
The first argument given is the integer matrix A.

The second argument given is the integer array B.

The third argument given is the integer array C.

The fourth argument given is the integer array D.

The fifth argument given is the integer array E.

(B[i], C[i]) represents the top left corner of the i'th query.

(D[i], E[i]) represents the bottom right corner of the i'th query.

### Output Format
Return the submatrix sum (% 10^9 + 7) for each query in the form of an integer array.

### Constraints
1 <= N, M <= 1000

-100000 <= A[i] <= 100000

1 <= Q <= 100000

1 <= B[i] <= D[i] <= N

1 <= C[i] <= E[i] <= M

## For Example
### Input 1:
    A = [   [1, 2, 3]
            [4, 5, 6]
            [7, 8, 9]   ]
            
    B = [1, 2]
    
    C = [1, 2]
    
    D = [2, 3]
    
    E = [2, 3]

### Output 1:
    [12, 28]

### Input 2:
    A = [   [5, 17, 100, 11]
            [0, 0,  2,   8]    ]
    
    B = [1, 1]
    
    C = [1, 4]
    
    D = [2, 2]
    
    E = [2, 4]

### Output 2:
    [22, 19]

```python
def solve(a, B, C, D, E):
        q = len(B)
        t = 0
        m = len(a)
        mod = 1000000007
        n = len(a[0])
        res = []
        prefix_sum = [[0 for i in range(n)] for i in range(m)] 
        for i in range(n): 
            prefix_sum[0][i] = a[0][i] % mod 
        for i in range(1, m): 
            for j in range(n): 
                prefix_sum[i][j] = (a[i][j] + prefix_sum[i - 1][j]) % mod
        for i in range(m): 
            for j in range(1, n): 
                prefix_sum[i][j] += prefix_sum[i][j - 1] % mod
        while t < q:
            tli, tlj = B[t], C[t]
            bri, brj = D[t], E[t]
            tli -= 1
            tlj -= 1
            bri -= 1
            brj -= 1
            sum = prefix_sum[bri][brj] 
            if (tli > 0): 
                sum = sum - prefix_sum[tli - 1][brj] % mod
            if (tlj > 0): 
                sum = sum - prefix_sum[bri][tlj - 1] % mod
            if (tli > 0 and tlj > 0): 
                sum = sum + prefix_sum[tli - 1][tlj - 1] % mod
    
            res.append(sum%mod)
            t += 1
    
        return res
    
A = [   [5, 17, 100, 11],
            [0, 0,  2,   8]    ]
B = [1, 1]
C = [1, 4]
D = [2, 2]
E = [2, 4]

print(solve(A,B,C,D,E))
```
